```python
data_cleaned = retail_data.drop_duplicates()

data_cleaned = data_cleaned.dropna(subset=['CustomerID'])

data_cleaned.isnull().sum()
```

```python
data_cleaned.duplicated().sum()
```

```python
data_cleaned.head()
```

```python
data_cleaned['TotalAmount'] = data_cleaned['Quantity'] * data_cleaned['UnitPrice']

data_cleaned.head()
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])

data_cleaned.head()
```

```python
data_cleaned.info()
```

```python
data_cleaned['Year'] = data_cleaned['InvoiceDate'].dt.year
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.month
data_cleaned['Day'] = data_cleaned['InvoiceDate'].dt.day
data_cleaned['Hour'] = data_cleaned['InvoiceDate'].dt.hour

data_cleaned.head()
```

```python
sales_by_country = data_cleaned.groupby('Country')['TotalAmount'].sum().reset_index()

sales_by_country = sales_by_country.sort_values(by='TotalAmount', ascending=False)

sales_by_country.head()
```

```python
sales_by_year = data_cleaned.groupby('Year')['TotalAmount'].sum().reset_index()

sales_by_year.head()
```

```python
data_cleaned['Semester'] = data_cleaned['Month'].apply(lambda x: 'Semester 1' if x <= 6 else 'Semester 2')

data_cleaned.head()
```

```python
sales_by_semester = data_cleaned.groupby(['Year', 'Semester'])['TotalAmount'].sum().reset_index()

sales_by_semester.head()
```

```python
data_cleaned['Trimester'] = data_cleaned['Month'].apply(lambda x: 'Trimester 1' if x <= 3 else 'Trimester 2' if x <= 6 else 'Trimester 3' if x <= 9 else 'Trimester 4')

data_cleaned.head()
```

```python
sales_by_trimester = data_cleaned.groupby(['Year', 'Trimester'])['TotalAmount'].sum().reset_index()

sales_by_trimester.head()
```

```python
sales_by_Month = data_cleaned.groupby(['Month'])['TotalAmount'].sum().reset_index()

sales_by_Month
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 6))
gs = GridSpec(2, 2, width_ratios=[1, 2], height_ratios=[2, 1])  # 2 filas, 2 columnas

# Subplot superior que ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0, :])
ax1.plot([1, 2, 3], label="Grfico 1")

# Subplot inferior izquierdo
ax2 = fig.add_subplot(gs[1, 0])
ax2.scatter([1, 2, 3], [1, 2, 3])

# Subplot inferior derecho
ax3 = fig.add_subplot(gs[1, 1])
ax3.bar([1, 2, 3], [3, 2, 1])

plt.tight_layout()
plt.show()
```
